### Analysis Steps
## 1. Data Loading and Cleaning

```python
### Import Libraries and Load Data
import pandas as pd
```

```python
# Load the data
df = pd.read_csv('../Volume-Analysis-and-Market-Trends/data/sp500_data.csv')
```

### Clean the Data

```python
#check missing values
df[df.isna().any(axis = 1)]
```

```python
# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

## 2. Exploratory Data Analysis (EDA)

```python
import plotly.express as px

volumn_trends = df.groupby('Date')['Volume'].sum().reset_index()

# Create an interactive plot using Plotly
fig = px.line(volumn_trends , x = 'Date' , y = 'Volume' , title = 'Overall Trading Volume Trends' , line_shape= 'linear')
fig.update_traces(line=dict(color = 'royalblue'))
fig.update_layout(xaxis_title= 'Date' , yaxis_title='Volume')
fig.show()
```
